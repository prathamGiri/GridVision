```python
import pandas as pd 
import numpy as np
```

```python
solar_data = pd.read_csv("Solar Data/Daily Data/Solar 2020 pavagada daily data.csv")
gen_data = pd.read_csv("Gen Data/FORTUM SOLAR 250MW Gen Data.csv")
```

```python
solar_data.head()
```

```python
solar_data.count()
```

```python
gen_data.head()
```

```python
# Find days present in solar_data but missing in gen_data
solar_days = set(zip(solar_data['Month'], solar_data['Day']))
gen_days = set(zip(gen_data['Month'], gen_data['Day']))

missing_in_gen = solar_days - gen_days
missing_in_solar = gen_days - solar_days

print("Days present in solar_data but missing in gen_data:")
for month, day in sorted(missing_in_gen):
    print(f"Month: {month}, Day: {day}")

print("\nDays present in gen_data but missing in solar_data:")
for month, day in sorted(missing_in_solar):
    print(f"Month: {month}, Day: {day}")
```

```python
merged_df = pd.merge(solar_data, gen_data, on=['Month', 'Day'], how='inner')
merged_df.head()
```

```python
merged_df.count()
```

```python
merged_df.to_csv("Complete Data/FORTUM SOLAR 250MW Complete Data.csv", index=False)
```

```python
daily_df = pd.read_csv("complete data/FORTUM SOLAR 250MW Complete Data.csv")
daily_df.head()
```

```python
hourly_df = pd.read_csv("solar data/Hourly Data/Solar 2020 pavagada hourly data.csv")
hourly_df.head(24)
```

```python
hourly_df = hourly_df.drop(columns = ["Year", 'Minute','Asymmetry','Alpha','Dew Point','Fill Flag','Ozone','Solar Zenith Angle','Surface Albedo', 'Pressure', 'Precipitable Water', 'Wind Direction', 'Wind Speed '])
```

```python
feature_order = ['Month', 'Day', 'Hour', 'Temperature', 'Cloud Type', 'Relative Humidity',
       'Aerosol Optical Depth', 'Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI',
       'DHI', 'DNI', 'GHI']
hourly_df = hourly_df[feature_order]
hourly_df.head(24)
```

```python
one_day = hourly_df[hourly_df['Day'] == 30][hourly_df['Month'] == 12]
```

```python
one_day["Station"] = "FORTUM SOLAR"
one_day["Operational Capacity(MW)"] = 250
one_day["Latitude"] = 14.23	
one_day["Longitude"] = 77.46
```

```python
one_day.to_csv("../backend/API data/hourly/FORTUM SOLAR_hourly_api_data.csv", index=False)
```
