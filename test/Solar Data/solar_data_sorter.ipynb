```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Hourly Data/Solar 2020 pavagada hourly data.csv')
df.head()
```

```python
cols_to_drop = ['Hour', 'Minute','Year','Alpha', 'Dew Point', 'Fill Flag', 'Asymmetry', 'Surface Albedo', 'Ozone', 'Pressure', 'Precipitable Water', 'Solar Zenith Angle', 'Wind Speed', 'Wind Direction']
df = df.drop(columns=cols_to_drop)
```

```python
df.head(24)
```

```python
df2 = df.copy()
```

```python
df2 = df2.drop(columns=['Temperature', 'Relative Humidity', 'Cloud Type', 'Aerosol Optical Depth'])
```

```python
df1 = df.copy()
df1 = df1.drop(columns=['Clearsky DNI', 'DNI', 'DHI', 'GHI', 'Clearsky GHI', 'Clearsky DHI'])
df1.head()
```

```python
df1_daily_avg = df1.groupby(['Month', 'Day'])[['Temperature', 'Cloud Type', 'Relative Humidity','Aerosol Optical Depth']].mean().reset_index()
df1_daily_avg.head()
```

```python
df_daily_avg_nonzero = (
    df2.replace(0, np.nan)
      .groupby(['Month', 'Day'])
      .mean()
      .reset_index()
)
df_daily_avg_nonzero.head()
```

```python
df_merged = pd.merge(df1_daily_avg, df_daily_avg_nonzero, on=['Month', 'Day'])
df_merged.head()
```

```python
# Round up the 'Cloud Type' column in df_merged to the nearest integer
df_merged['Cloud Type'] = np.ceil(df_merged['Cloud Type']).astype(int)
df_merged.head()
```

```python
df_merged.to_csv('Daily Data/Solar 2020 pavagada daily data.csv', index=False)
```
